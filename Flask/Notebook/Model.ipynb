```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import GridSearchCV

import pickle
```

```python
url="data/Sleep_health_and_lifestyle_dataset.csv"
data=pd.read_csv(url,encoding='latin-1')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data["Sleep Disorder"].value_counts()
```

```python
data.columns = data.columns.str.replace(' ', '_')
```

```python
data.head()
```

```python
data.columns
```

```python
numeric_data = data.select_dtypes(include=['number'])
numeric_data.corr()['Sleep_Duration']
```

```python
data.head()
```

```python
data=data.drop(columns=['Person_ID', 'Sleep_Disorder'])
```

```python
data.head()
```

```python
data = pd.get_dummies(data, columns=['Gender'], drop_first=False)
```

```python
data.head()
```

```python
data['Gender_Female'] = data['Gender_Female'].astype(int)
data['Gender_Male'] = data['Gender_Male'].astype(int)
```

```python
data.head()
```

```python
# Ensure no leading/trailing spaces in the BMI_Category column
data['BMI_Category'] = data['BMI_Category'].str.strip()

# Apply one-hot encoding
bmi_encoded = pd.get_dummies(data['BMI_Category'], prefix='BMI')

# Concatenate the encoded columns back to the dataset
data = pd.concat([data, bmi_encoded], axis=1)

# Drop the original 'BMI_Category' column
data.drop('BMI_Category', axis=1, inplace=True)
```

```python
data.head()
```

```python
bmi_columns = ['BMI_Normal','BMI_Normal Weight', 'BMI_Obese', 'BMI_Overweight']
for col in bmi_columns:
    data[col] = data[col].astype(int)
```

```python
data.head()
```

```python
data.rename(columns=lambda x: x.replace(" ", "_"), inplace=True)
```

```python
data.head()
```

```python
data=data.drop(columns=['Occupation'])
```

```python
data.head()
```

```python
data.info()
```

```python
data[['Systolic_BP', 'Diastolic_BP']] = data['Blood_Pressure'].str.split('/', expand=True)

# Convert the new columns to integers
data['Systolic_BP'] = data['Systolic_BP'].astype(int)
data['Diastolic_BP'] = data['Diastolic_BP'].astype(int)

# Drop the original 'Blood_Pressure' column (optional)
data.drop('Blood_Pressure', axis=1, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
numeric_data = data.select_dtypes(include=['number'])
numeric_data.corr()['Sleep_Duration']
```

```python
data.shape
```

```python
X=data.drop('Sleep_Duration',axis=1)
y=data['Sleep_Duration']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape,X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train,y_train)
    acc=model.score(X_test,y_test)
    print(str(model)+ '-->' + str(acc))
```

```python
lr=LinearRegression()
model_acc(lr)

lasso=Lasso()
model_acc(lasso)

dt = DecisionTreeRegressor()
model_acc(dt)

rf = RandomForestRegressor()
model_acc(rf)
```

```python

parameters = {'n_estimators':[10, 50, 100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(X_train, y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test, y_test)
```

```python
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
